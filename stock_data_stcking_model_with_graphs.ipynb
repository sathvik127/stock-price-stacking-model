```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
from sklearn.inspection import permutation_importance
from numpy import nan as npNaN
import joblib
import yfinance as yf
from datetime import datetime
import pandas_ta as ta
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.linear_model import Ridge, HuberRegressor, LinearRegression, RANSACRegressor
import xgboost as xgb

from scipy.stats import skew, kurtosis, zscore
import seaborn as sns
import matplotlib.pyplot as plt

WINDOW_SIZE = 20
FORECAST_HORIZON = 14
REQUIRED_HOLDOUT_DAYS = WINDOW_SIZE + FORECAST_HORIZON # 34

def fetch_stock_data(ticker, start_date, end_date):
    print(f"Fetching stock data for {ticker} from {start_date} to {end_date}...")
    df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=False)
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)
    return df

def add_technical_and_statistical_features(df):
    df['High'] = pd.to_numeric(df['High'], errors='coerce')
    df['Low'] = pd.to_numeric(df['Low'], errors='coerce')
    df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
    df['Returns'] = df['Close'].pct_change()

    df['SMA_50'] = ta.sma(df['Close'], 50)
    df['SMA_200'] = ta.sma(df['Close'], 200)
    df['RSI'] = ta.rsi(df['Close'], 14)
    macd = ta.macd(df['Close']); df['MACD'] = macd['MACD_12_26_9']; df['MACD_signal'] = macd['MACDs_12_26_9']
    bbands = ta.bbands(df['Close']); df['BB_upper'] = bbands['BBU_5_2.0']; df['BB_lower'] = bbands['BBL_5_2.0']
    df['ATR'] = ta.atr(df['High'], df['Low'], df['Close'], 14)

    df['BB_width'] = (df['BB_upper'] - df['BB_lower']) / df['Close']
    df['Volatility_10d'] = df['Returns'].rolling(10).std()
    df['ROC_10'] = ta.roc(df['Close'], 10)
    df['Momentum'] = df['Close'] - df['Close'].shift(4)
    df['Skew_10d'] = df['Returns'].rolling(10).skew()
    df['Kurtosis_10d'] = df['Returns'].rolling(10).kurt()

    df.dropna(inplace=True)
    return df

def create_features_and_targets(df, window_size, forecast_horizon):
    features = [
        'Close', 'SMA_50', 'SMA_200', 'RSI', 'MACD', 'MACD_signal',
        'BB_upper', 'BB_lower', 'ATR', 'BB_width', 'Volatility_10d',
        'ROC_10', 'Momentum', 'Skew_10d', 'Kurtosis_10d'
    ]
    X, y, dates, last_known_prices = [], [], [], []
    for i in range(len(df) - window_size - forecast_horizon):
        feature_values = df[features].iloc[i : i + window_size].values.flatten()
        last_close = df['Close'].iloc[i + window_size - 1]
        target_values = df['Close'].iloc[i + window_size : i + window_size + forecast_horizon].values
        target_changes = target_values - last_close
        if not np.isnan(feature_values).any() and not np.isnan(target_changes).any():
            X.append(feature_values)
            y.append(target_changes)
            dates.append(df.index[i + window_size])
            last_known_prices.append(last_close)
    return np.array(X), np.array(y), features, np.array(dates), np.array(last_known_prices)

def create_single_feature_row_for_prediction(df, window_size):
    features = [
        'Close', 'SMA_50', 'SMA_200', 'RSI', 'MACD', 'MACD_signal',
        'BB_upper', 'BB_lower', 'ATR', 'BB_width', 'Volatility_10d',
        'ROC_10', 'Momentum', 'Skew_10d', 'Kurtosis_10d'
    ]
    if len(df) < window_size:
        raise ValueError(f"Not enough data for prediction. Need {window_size} days, have {len(df)}.")
    return df[features].iloc[-window_size:].values.flatten().reshape(1, -1)

def get_model_for_group(group_type):
    if group_type == 'linear': return Ridge(alpha=1.0)
    if group_type == 'robust': return RANSACRegressor(estimator=LinearRegression(), max_trials=100, min_samples=0.9)
    return xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, random_state=42)

def main():
    ticker = "COALINDIA.NS"
    start_date = "2013-01-01"
    end_date = datetime.today().strftime('%Y-%m-%d')

    full_df = fetch_stock_data(ticker, start_date, end_date)
    full_df = add_technical_and_statistical_features(full_df)

    print(f"\n--- Isolating First {REQUIRED_HOLDOUT_DAYS} Trading Days of 2022 for Hold-Out Test ---")
    holdout_candidate_df = full_df.loc['2022-01-01':].copy()

    if len(holdout_candidate_df) < REQUIRED_HOLDOUT_DAYS:
        print(f"Warning: Not enough trading data in 2022 for the hold-out test. Skipping.")
        use_holdout = False
        df = full_df
    else:
        use_holdout = True
        holdout_df = holdout_candidate_df.iloc[:REQUIRED_HOLDOUT_DAYS]
        holdout_start = holdout_df.index.min().strftime('%Y-%m-%d'); holdout_end = holdout_df.index.max().strftime('%Y-%m-%d')
        print(f"Successfully isolated {len(holdout_df)} days from {holdout_start} to {holdout_end}.")

        X_holdout = create_single_feature_row_for_prediction(holdout_df.iloc[:WINDOW_SIZE], WINDOW_SIZE)
        y_holdout_actual_df = holdout_df.iloc[WINDOW_SIZE : WINDOW_SIZE + FORECAST_HORIZON]
        y_holdout_actual_prices = y_holdout_actual_df['Close'].values
        holdout_target_dates = y_holdout_actual_df.index

        df = full_df.drop(holdout_df.index)
        print(f"Main dataset now has {len(df)} days.")

    X_tomorrow = create_single_feature_row_for_prediction(df, WINDOW_SIZE)
    X, y, base_features, dates, last_known_prices = create_features_and_targets(df, WINDOW_SIZE, FORECAST_HORIZON)
    print(f"\n--- VERIFYING SHAPES ---\nX shape: {X.shape}\ny shape: {y.shape}\n------------------------")

    # total = len(X)
    # train_end = int(total * 0.8)
    # X_train, y_train = X[:train_end], y[:train_end]
    # X_test, y_test, last_known_prices_test = X[train_end:], y[train_end:], last_known_prices[train_end:]



    print("\n--- Splitting data into 70% Train, 15% Validation, 15% Test ---")
    total_samples = len(X)
    train_end = int(total_samples * 0.70)
    val_end = train_end + int(total_samples * 0.15)

    X_train, y_train = X[:train_end], y[:train_end]
    X_val, y_val = X[train_end:val_end], y[train_end:val_end]
    X_test, y_test = X[val_end:], y[val_end:]

    last_known_prices_test = last_known_prices[val_end:]







    mask_train = ~np.isnan(X_train).any(axis=1) & ~np.isnan(y_train).any(axis=1)
    X_train, y_train = X_train[mask_train], y_train[mask_train]
    mask_test = ~np.isnan(X_test).any(axis=1) & ~np.isnan(y_test).any(axis=1)
    X_test, y_test, last_known_prices_test = X_test[mask_test], y_test[mask_test], last_known_prices_test[mask_test]

    all_final_preds_test, all_final_preds_tomorrow = [], []
    if use_holdout: all_final_preds_holdout = []

    print(f"\n====================== Building {FORECAST_HORIZON} Stacking Models ======================\n")
    for target_idx in range(y_train.shape[1]):
        print(f"--- Training Model for Target Day {target_idx + 1}/{FORECAST_HORIZON} ---")
        group_preds_train, group_preds_test, group_preds_tomorrow = [], [], []
        group_preds_val = []
        if use_holdout: group_preds_holdout = []

        for group_type in ['linear', 'robust', 'nonlinear']:
            pipe = make_pipeline(StandardScaler(), get_model_for_group(group_type))
            pipe.fit(X_train, y_train[:, target_idx])

            group_preds_val.append(pipe.predict(X_val).reshape(-1, 1))

            group_preds_train.append(pipe.predict(X_train).reshape(-1, 1))
            group_preds_test.append(pipe.predict(X_test).reshape(-1, 1))
            group_preds_tomorrow.append(pipe.predict(X_tomorrow).reshape(-1, 1))
            if use_holdout and X_holdout.size > 0: group_preds_holdout.append(pipe.predict(X_holdout).reshape(-1, 1))

        stacked_X_val = np.hstack(group_preds_val); stacked_X_test = np.hstack(group_preds_test)
        stacked_X_tomorrow = np.hstack(group_preds_tomorrow)
        if use_holdout and X_holdout.size > 0: stacked_X_holdout = np.hstack(group_preds_holdout)

        stack_model = Ridge(); stack_model.fit(stacked_X_val, y_val[:, target_idx])
        all_final_preds_test.append(stack_model.predict(stacked_X_test))
        all_final_preds_tomorrow.append(stack_model.predict(stacked_X_tomorrow))
        if use_holdout and X_holdout.size > 0: all_final_preds_holdout.append(stack_model.predict(stacked_X_holdout))

    final_preds_test_array = np.array(all_final_preds_test).T
    final_preds_tomorrow_array = np.array(all_final_preds_tomorrow).T
    print("\n\n")

    print("====================== Overall Model Evaluation (on Test Set) ======================")
    predicted_prices_test = last_known_prices_test[:, np.newaxis] + final_preds_test_array
    actual_prices_test = last_known_prices_test[:, np.newaxis] + y_test

    mae = mean_absolute_error(actual_prices_test, predicted_prices_test)
    mape = mean_absolute_percentage_error(actual_prices_test, predicted_prices_test) * 100
    accuracy = 100 - mape
    print(f"Overall MAE on Price Levels: {mae:.4f}")
    print(f"Overall MAPE: {mape:.2f}% | Overall Accuracy: {accuracy:.2f}%")

    if use_holdout and X_holdout.size > 0:
        print("\n\n====================== Hold-Out Set Validation (Early 2022) ======================")
        final_preds_holdout_array = np.array(all_final_preds_holdout).T.squeeze()
        last_close_before_holdout = holdout_df.iloc[WINDOW_SIZE - 1]['Close']
        predicted_prices_holdout = last_close_before_holdout + final_preds_holdout_array

        comparison_df = pd.DataFrame({
            'Actual Close': y_holdout_actual_prices,
            'Predicted Close': predicted_prices_holdout
        }, index=holdout_target_dates)

        print("\n--- Comparison of Predicted vs. Actual Close Prices (Early 2022) ---")
        print(comparison_df.round(2))

        holdout_mae = mean_absolute_error(comparison_df['Actual Close'], comparison_df['Predicted Close'])
        print(f"\nHold-Out MAE (on Price Levels): {holdout_mae:.4f}")

        plt.figure(figsize=(14, 7))
        plt.plot(comparison_df.index, comparison_df['Actual Close'], 'o-', label='Actual Close', color='blue')
        plt.plot(comparison_df.index, comparison_df['Predicted Close'], 'o--', label='Predicted Close', color='red')
        plt.title(f'Hold-Out Set Performance for {ticker} (Early 2022) - 14-Day Horizon', fontsize=16)
        plt.xlabel('Date'); plt.ylabel('Price')
        plt.legend(); plt.grid(True, linestyle='--'); plt.tight_layout()
        plt.show()

    print("\n\n====================== Full 14-Day Forecast (From Today) ======================")
    predicted_changes_tomorrow = final_preds_tomorrow_array.squeeze()
    last_historical_close = df['Close'].iloc[-1]
    last_historical_date = df.index[-1]

    predicted_prices_tomorrow = last_historical_close + predicted_changes_tomorrow

    forecast_dates = pd.to_datetime([last_historical_date + pd.Timedelta(days=i) for i in range(1, FORECAST_HORIZON + 1)])
    forecast_series = pd.Series(predicted_prices_tomorrow, index=forecast_dates)

    print("\n--- Forecasted Close Prices for the Next 14 Days ---")
    print(forecast_series.round(2))

    historical_data = df['Close'].iloc[-60:].copy()
    plt.figure(figsize=(15, 7))
    plt.plot(historical_data.index, historical_data, label='Historical Close', color='blue')
    plt.plot(forecast_series.index, forecast_series, 'o--', label=f'Predicted Close ({FORECAST_HORIZON} Days)', color='orange')
    plt.title(f'{FORECAST_HORIZON}-Day Stock Price Forecast for {ticker}', fontsize=16)
    plt.xlabel('Date', fontsize=12); plt.ylabel('Price', fontsize=12)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.legend(); plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
from sklearn.inspection import permutation_importance
from numpy import nan as npNaN
import joblib
import yfinance as yf
from datetime import datetime
import pandas_ta as ta
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.linear_model import Ridge, HuberRegressor, LinearRegression, RANSACRegressor
import xgboost as xgb

from scipy.stats import skew, kurtosis, zscore
import seaborn as sns
import matplotlib.pyplot as plt

WINDOW_SIZE = 20
FORECAST_HORIZON = 14
REQUIRED_HOLDOUT_DAYS = WINDOW_SIZE + FORECAST_HORIZON # 34

def fetch_stock_data(ticker, start_date, end_date):
    print(f"Fetching stock data for {ticker} from {start_date} to {end_date}...")
    df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=False)
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)
    return df

def add_technical_and_statistical_features(df):
    df['High'] = pd.to_numeric(df['High'], errors='coerce')
    df['Low'] = pd.to_numeric(df['Low'], errors='coerce')
    df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
    df['Returns'] = df['Close'].pct_change()

    df['SMA_50'] = ta.sma(df['Close'], 50)
    df['SMA_200'] = ta.sma(df['Close'], 200)
    df['RSI'] = ta.rsi(df['Close'], 14)
    macd = ta.macd(df['Close']); df['MACD'] = macd['MACD_12_26_9']; df['MACD_signal'] = macd['MACDs_12_26_9']
    bbands = ta.bbands(df['Close']); df['BB_upper'] = bbands['BBU_5_2.0']; df['BB_lower'] = bbands['BBL_5_2.0']
    df['ATR'] = ta.atr(df['High'], df['Low'], df['Close'], 14)

    df['BB_width'] = (df['BB_upper'] - df['BB_lower']) / df['Close']
    df['Volatility_10d'] = df['Returns'].rolling(10).std()
    df['ROC_10'] = ta.roc(df['Close'], 10)
    df['Momentum'] = df['Close'] - df['Close'].shift(4)
    df['Skew_10d'] = df['Returns'].rolling(10).skew()
    df['Kurtosis_10d'] = df['Returns'].rolling(10).kurt()

    df.dropna(inplace=True)
    return df

def create_features_and_targets(df, window_size, forecast_horizon):
    features = [
        'Close', 'SMA_50', 'SMA_200', 'RSI', 'MACD', 'MACD_signal',
        'BB_upper', 'BB_lower', 'ATR', 'BB_width', 'Volatility_10d',
        'ROC_10', 'Momentum', 'Skew_10d', 'Kurtosis_10d'
    ]
    X, y, dates, last_known_prices = [], [], [], []
    for i in range(len(df) - window_size - forecast_horizon):
        feature_values = df[features].iloc[i : i + window_size].values.flatten()
        last_close = df['Close'].iloc[i + window_size - 1]
        target_values = df['Close'].iloc[i + window_size : i + window_size + forecast_horizon].values
        target_changes = target_values - last_close
        if not np.isnan(feature_values).any() and not np.isnan(target_changes).any():
            X.append(feature_values)
            y.append(target_changes)
            dates.append(df.index[i + window_size])
            last_known_prices.append(last_close)
    return np.array(X), np.array(y), features, np.array(dates), np.array(last_known_prices)

def create_single_feature_row_for_prediction(df, window_size):
    features = [
        'Close', 'SMA_50', 'SMA_200', 'RSI', 'MACD', 'MACD_signal',
        'BB_upper', 'BB_lower', 'ATR', 'BB_width', 'Volatility_10d',
        'ROC_10', 'Momentum', 'Skew_10d', 'Kurtosis_10d'
    ]
    if len(df) < window_size:
        raise ValueError(f"Not enough data for prediction. Need {window_size} days, have {len(df)}.")
    return df[features].iloc[-window_size:].values.flatten().reshape(1, -1)

def get_model_for_group(group_type):
    if group_type == 'linear': return Ridge(alpha=1.0)
    if group_type == 'robust': return RANSACRegressor(estimator=LinearRegression(), max_trials=100, min_samples=0.9)
    return xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, random_state=42)

def main():
    ticker = "COALINDIA.NS"
    start_date = "2013-01-01"
    end_date = datetime.today().strftime('%Y-%m-%d')

    full_df = fetch_stock_data(ticker, start_date, end_date)
    full_df = add_technical_and_statistical_features(full_df)

    print(f"\n--- Isolating First {REQUIRED_HOLDOUT_DAYS} Trading Days of 2022 for Hold-Out Test ---")
    holdout_candidate_df = full_df.loc['2022-01-01':].copy()

    if len(holdout_candidate_df) < REQUIRED_HOLDOUT_DAYS:
        print(f"Warning: Not enough trading data in 2022 for the hold-out test. Skipping.")
        use_holdout = False
        df = full_df
    else:
        use_holdout = True
        holdout_df = holdout_candidate_df.iloc[:REQUIRED_HOLDOUT_DAYS]
        holdout_start = holdout_df.index.min().strftime('%Y-%m-%d'); holdout_end = holdout_df.index.max().strftime('%Y-%m-%d')
        print(f"Successfully isolated {len(holdout_df)} days from {holdout_start} to {holdout_end}.")

        X_holdout = create_single_feature_row_for_prediction(holdout_df.iloc[:WINDOW_SIZE], WINDOW_SIZE)
        y_holdout_actual_df = holdout_df.iloc[WINDOW_SIZE : WINDOW_SIZE + FORECAST_HORIZON]
        y_holdout_actual_prices = y_holdout_actual_df['Close'].values
        holdout_target_dates = y_holdout_actual_df.index

        df = full_df.drop(holdout_df.index)
        print(f"Main dataset now has {len(df)} days.")

    X_tomorrow = create_single_feature_row_for_prediction(df, WINDOW_SIZE)
    X, y, base_features, dates, last_known_prices = create_features_and_targets(df, WINDOW_SIZE, FORECAST_HORIZON)
    print(f"\n--- VERIFYING SHAPES ---\nX shape: {X.shape}\ny shape: {y.shape}\n------------------------")

    print("\n--- Splitting data into 70% Train, 15% Validation, 15% Test ---")
    total_samples = len(X)
    train_end = int(total_samples * 0.70)
    val_end = train_end + int(total_samples * 0.15)
    X_train, y_train = X[:train_end], y[:train_end]
    X_val, y_val = X[train_end:val_end], y[train_end:val_end]
    X_test, y_test = X[val_end:], y[val_end:]
    last_known_prices_test = last_known_prices[val_end:]

    all_final_preds_test, all_final_preds_tomorrow = [], []
    if use_holdout: all_final_preds_holdout = []

    print(f"\n====================== Building {FORECAST_HORIZON} Stacking Models ======================\n")
    for target_idx in range(y_train.shape[1]):
        print(f"--- Training Model for Target Day {target_idx + 1}/{FORECAST_HORIZON} ---")
        group_preds_val = []
        group_preds_test = []
        group_preds_tomorrow = []
        if use_holdout: group_preds_holdout = []

        for group_type in ['linear', 'robust', 'nonlinear']:
            pipe = make_pipeline(StandardScaler(), get_model_for_group(group_type))
            pipe.fit(X_train, y_train[:, target_idx])
            group_preds_val.append(pipe.predict(X_val).reshape(-1, 1))
            group_preds_test.append(pipe.predict(X_test).reshape(-1, 1))
            group_preds_tomorrow.append(pipe.predict(X_tomorrow).reshape(-1, 1))
            if use_holdout and X_holdout.size > 0: group_preds_holdout.append(pipe.predict(X_holdout).reshape(-1, 1))

        stacked_X_val = np.hstack(group_preds_val); stacked_X_test = np.hstack(group_preds_test)
        stacked_X_tomorrow = np.hstack(group_preds_tomorrow)
        if use_holdout and X_holdout.size > 0: stacked_X_holdout = np.hstack(group_preds_holdout)

        stack_model = Ridge(); stack_model.fit(stacked_X_val, y_val[:, target_idx])
        all_final_preds_test.append(stack_model.predict(stacked_X_test))
        all_final_preds_tomorrow.append(stack_model.predict(stacked_X_tomorrow))
        if use_holdout and X_holdout.size > 0: all_final_preds_holdout.append(stack_model.predict(stacked_X_holdout))

    final_preds_test_array = np.array(all_final_preds_test).T
    final_preds_tomorrow_array = np.array(all_final_preds_tomorrow).T
    print("\n\n")

    print("====================== Overall Model Evaluation (on Test Set) ======================")
    predicted_prices_test = last_known_prices_test[:, np.newaxis] + final_preds_test_array
    actual_prices_test = last_known_prices_test[:, np.newaxis] + y_test
    mae = mean_absolute_error(actual_prices_test, predicted_prices_test)
    mape = mean_absolute_percentage_error(actual_prices_test, predicted_prices_test) * 100
    accuracy = 100 - mape
    print(f"Overall MAE on Price Levels: {mae:.4f}")
    print(f"Overall MAPE: {mape:.2f}% | Overall Accuracy: {accuracy:.2f}%")

    if use_holdout and X_holdout.size > 0:
        pass

    print("\n\n====================== Full 14-Day Forecast (From Today) ======================")
    predicted_changes_tomorrow = final_preds_tomorrow_array.squeeze()
    last_historical_close = df['Close'].iloc[-1]
    last_historical_date = df.index[-1]
    predicted_prices_tomorrow = last_historical_close + predicted_changes_tomorrow
    forecast_dates = pd.to_datetime([last_historical_date + pd.Timedelta(days=i) for i in range(1, FORECAST_HORIZON + 1)])
    forecast_series = pd.Series(predicted_prices_tomorrow, index=forecast_dates)
    print("\n--- Forecasted Close Prices for the Next 14 Days ---")
    print(forecast_series.round(2))


    print("\n\n========================== Trading Signal ==========================")


    TRADE_THRESHOLD = 1.0


    predicted_price_end_of_horizon = forecast_series.iloc[-1]


    predicted_return_percent = ((predicted_price_end_of_horizon / last_historical_close) - 1) * 100

    if predicted_return_percent > TRADE_THRESHOLD:
        signal = "BUY"
    elif predicted_return_percent < -TRADE_THRESHOLD:
        signal = "SELL"
    else:
        signal = "HOLD"

    print(f"Last Known Close Price: {last_historical_close:.2f}")
    print(f"Predicted Price in {FORECAST_HORIZON} days: {predicted_price_end_of_horizon:.2f}")
    print(f"Predicted Return: {predicted_return_percent:.2f}%")
    print("------------------------------------------------------------------")
    print(f"DECISION: *** {signal} ***")
    print("------------------------------------------------------------------")
    print("\nDisclaimer: This is for educational purposes only and is not financial advice.")
    print("All trading involves risk. Do your own research.")

    historical_data = df['Close'].iloc[-60:].copy()
    plt.figure(figsize=(15, 7))
    plt.plot(historical_data.index, historical_data, label='Historical Close', color='blue')
    plt.plot(forecast_series.index, forecast_series, 'o--', label=f'Predicted Close ({FORECAST_HORIZON} Days)', color='orange')
    plt.title(f'{FORECAST_HORIZON}-Day Stock Price Forecast for {ticker}', fontsize=16)
    plt.xlabel('Date', fontsize=12); plt.ylabel('Price', fontsize=12)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.legend(); plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.inspection import permutation_importance
from numpy import nan as npNaN
import joblib
import yfinance as yf
from datetime import datetime
import pandas_ta as ta
import numpy as np

from sklearn.ensemble import HistGradientBoostingRegressor

from sklearn.linear_model import Ridge, HuberRegressor
import xgboost as xgb

from scipy.stats import skew, kurtosis, zscore
import seaborn as sns
import matplotlib.pyplot as plt


def fetch_stock_data(ticker, start_date, end_date):
    print(f"Fetching stock data for {ticker} from {start_date} to {end_date}...")
    data = yf.download(ticker, start=start_date, end=end_date, auto_adjust=False)
    print(data.head())
    return data


def calculate_indicators(df):
    df['High'] = pd.to_numeric(df['High'], errors='coerce')
    df['Low'] = pd.to_numeric(df['Low'], errors='coerce')
    df['Close'] = pd.to_numeric(df['Close'], errors='coerce')

    df['SMA_50'] = ta.sma(df['Close'], 50)
    df['SMA_200'] = ta.sma(df['Close'], 200)
    df['RSI'] = ta.rsi(df['Close'], 14)

    macd = ta.macd(df['Close'])
    df['MACD'] = macd['MACD_12_26_9']
    df['MACD_signal'] = macd['MACDs_12_26_9']

    bbands = ta.bbands(df['Close'])
    df['BB_upper'] = bbands['BBU_5_2.0']
    df['BB_lower'] = bbands['BBL_5_2.0']

    df['Williams_R'] = ta.willr(df['High'], df['Low'], df['Close'], 14)
    df['ATR'] = ta.atr(df['High'], df['Low'], df['Close'], 14)

    return df


def full_feature_analysis_multitarget(df, features, target_columns):
    print("\n===== Feature Distribution Analysis =====")
    for col in features:
        if col in df.columns:
            series = df[col].dropna()
            print(f"\n---- {col} ----")
            print(f"Skewness: {skew(series):.2f}")
            print(f"Kurtosis: {kurtosis(series):.2f}")

    print("\n===== Correlation with Each Target =====")
    for target in target_columns:
        if target not in df.columns:
            continue
        selected_cols = [col for col in features if col in df.columns] + [target]
        corr_matrix = df[selected_cols].corr()
        if target in corr_matrix.columns:
            corr_with_target = corr_matrix[target].drop(target)
            print(f"\n--- Correlation with {target} ---")
            print(corr_with_target)

    print("\n===== Outlier Detection (Z-score > 3) =====")
    for col in features:
        if col in df.columns:
            z_scores = zscore(df[col].dropna())
            outliers = np.sum(np.abs(z_scores) > 3)
            print(f"{col}: {outliers} outliers")

    print("\n===== Full Correlation Heatmap =====")
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    sns.heatmap(df[numeric_cols].corr(), annot=False, cmap='coolwarm')
    plt.title("Correlation Heatmap")
    # plt.show()


# Feature engineering
def create_features(df, window_size=28):
    """
    Corrected version: Creates features from the last 28 days to predict
    the OHLC prices of the SINGLE NEXT DAY.
    """
    features = ['Close', 'SMA_50', 'SMA_200', 'RSI', 'MACD', 'MACD_signal', 'BB_upper', 'BB_lower', 'Williams_R', 'ATR']
    X = []
    y = []
    dates = []


    for i in range(len(df) - window_size - 1):
        feature_values = df[features].iloc[i:i + window_size].values.flatten()

        target_values = df[['Close', 'Adj Close', 'High', 'Low', 'Open']].iloc[i + window_size].values

        target_values = target_values.astype(float)
        X.append(feature_values)
        y.append(target_values)
        dates.append(df.index[i + window_size])

    return np.array(X), np.array(y), features, np.array(dates)


def create_single_feature_row_for_prediction(df, window_size=28):
    features = ['Close', 'SMA_50', 'SMA_200', 'RSI', 'MACD', 'MACD_signal', 'BB_upper', 'BB_lower', 'Williams_R', 'ATR']

    if len(df) < window_size:
        raise ValueError("Not enough data to create a prediction row.")

    feature_values = df[features].iloc[-window_size:].values.flatten()

    predicted_date = df.index[-1] + pd.Timedelta(days=1)

    return feature_values.reshape(1, -1), predicted_date



def group_features_per_target(df, features, targets, corr_threshold_linear=0.7, corr_threshold_robust=0.5, outlier_frac_threshold=0.05):
    feature_groups = {target: {'linear': [], 'robust': [], 'nonlinear': []} for target in targets}

    for target in targets:
        for feature in features:
            if feature not in df.columns or target not in df.columns:
                continue

            series = df[feature].dropna()
            tseries = df[target].dropna()
            min_len = min(len(series), len(tseries))
            series = series[-min_len:]
            tseries = tseries[-min_len:]

            corr = series.corr(tseries)
            sk = skew(series)
            kurt = kurtosis(series)
            z_scores = zscore(series)
            outlier_frac = np.sum(np.abs(z_scores) > 3) / len(z_scores)

            if abs(corr) >= corr_threshold_linear and abs(sk) < 1 and abs(kurt) < 3 and outlier_frac <= outlier_frac_threshold:
                feature_groups[target]['linear'].append(feature)
            elif abs(corr) >= corr_threshold_robust:
                feature_groups[target]['robust'].append(feature)
            else:
                feature_groups[target]['nonlinear'].append(feature)

    return feature_groups


from sklearn.linear_model import Ridge, HuberRegressor
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.ensemble import StackingRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler


def get_model_for_group(group_type):
    if group_type == 'linear':
        return Ridge(alpha=1.0)
    elif group_type == 'robust':
        return HuberRegressor(max_iter=1000)
    else:
        return xgb.XGBRegressor()

def main():
    all_metrics = []
    ticker = input("Enter the stock ticker symbol (e.g., COALINDIA.NS): ")
    start_date = "2013-01-01"
    end_date = datetime.today().strftime('%Y-%m-%d')


    def fetch_stock_data(ticker, start_date, end_date):
        print(f"Fetching stock data for {ticker} from {start_date} to {end_date}...")
        df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=False)
        if isinstance(df.columns, pd.MultiIndex):
            df.columns = df.columns.get_level_values(0)
        return df

    data = fetch_stock_data(ticker, start_date, end_date)
    file_path = f'{ticker}_stock_data-28days.csv'
    data.to_csv(file_path, index=True)
    print(f"Stock data saved to '{file_path}'.")

    df = pd.read_csv(file_path, index_col='Date', parse_dates=True)

    for col in ['Close', 'Adj Close', 'High', 'Low', 'Open']:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    df = calculate_indicators(df)

    updated_file_path = f'{ticker}_stock_data_with_indicators-28days.csv'
    df.to_csv(updated_file_path, index=True, date_format='%Y-%m-%d')
    print(f"Modified data with indicators saved to '{updated_file_path}'.")

    X_tomorrow, tomorrow_date = create_single_feature_row_for_prediction(df)
    X, y, base_features, dates = create_features(df)

    print("------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
    print("printing X and y",X.shape,y.shape)
    print(
        "------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

    total = len(X)
    train_end = int(total * 0.7)
    test1_end = train_end + int(total * 0.15)

    X_train, y_train, dates_train = X[:train_end], y[:train_end], dates[:train_end]
    X_test, y_test, dates_test = X[train_end:test1_end], y[train_end:test1_end], dates[train_end:test1_end]
    X_test2, y_test2, dates_test2 = X[test1_end:], y[test1_end:], dates[test1_end:]

    mask_train = ~np.isnan(X_train).any(axis=1);
    X_train, y_train = X_train[mask_train], y_train[mask_train]
    mask_test = ~np.isnan(X_test).any(axis=1);
    X_test, y_test = X_test[mask_test], y_test[mask_test]
    mask_test2 = ~np.isnan(X_test2).any(axis=1);
    X_test2, y_test2 = X_test2[mask_test2], y_test2[mask_test2]
    mask_tomorrow = ~np.isnan(X_tomorrow).any(axis=1);
    X_tomorrow = X_tomorrow[mask_tomorrow]

    target_cols = ['Close', 'Adj Close', 'High', 'Low', 'Open']
    full_feature_analysis_multitarget(df, base_features, target_cols)
    grouped = group_features_per_target(df, base_features, target_cols)

    tomorrow_predictions = {}

    print("\n====================== Final Model Evaluation ======================\n")

    for target_idx, target in enumerate(target_cols):
        print(f"\n Target: {target}")

        group_preds_val = []
        group_preds_test = []
        group_preds_tomorrow = []

        for group_type in ['linear', 'robust', 'nonlinear']:
            selected_features = grouped[target][group_type]
            if not selected_features: continue

            feature_indices = []
            for feat in selected_features:
                try:
                    base_idx = base_features.index(feat)

                    feature_indices.extend([base_idx * 28 + i for i in range(28)])
                except ValueError:
                    continue

            if not feature_indices: continue

            X_train_group = X_train[:, feature_indices]
            X_val_group = X_test[:, feature_indices]
            X_test_group = X_test2[:, feature_indices]
            X_tomorrow_group = X_tomorrow[:, feature_indices]

            model = get_model_for_group(group_type)
            pipe = make_pipeline(StandardScaler(), model)

            pipe.fit(X_train_group, y_train[:, target_idx])

            y_pred_val_group = pipe.predict(X_val_group)
            y_pred_test_group = pipe.predict(X_test_group)
            y_pred_tomorrow_group = pipe.predict(X_tomorrow_group)

            group_preds_val.append(y_pred_val_group.reshape(-1, 1))
            group_preds_test.append(y_pred_test_group.reshape(-1, 1))
            group_preds_tomorrow.append(y_pred_tomorrow_group.reshape(-1, 1))

        if not group_preds_val:
            print(f" No valid features found for target {target}. Skipping.")
            continue

        stacked_X_val = np.hstack(group_preds_val)
        stacked_X_test = np.hstack(group_preds_test)
        stacked_X_tomorrow = np.hstack(group_preds_tomorrow)

        stack_model = Ridge()
        stack_model.fit(stacked_X_val, y_test[:, target_idx])

        y_final_pred_test = stack_model.predict(stacked_X_test)
        y_final_pred_tomorrow = stack_model.predict(stacked_X_tomorrow)

        tomorrow_predictions[target] = y_final_pred_tomorrow[0]

        mae = mean_absolute_error(y_test2[:, target_idx], y_final_pred_test)
        mape = np.mean(np.abs((y_test2[:, target_idx] - y_final_pred_test) / y_test2[:, target_idx])) * 100
        accuracy = 100 - mape
        print(f" MAE: {mae:.4f} | MAPE: {mape:.2f}% | Accuracy: {accuracy:.2f}%")

        print("\n Predicted vs Actual values for the test period:")
        for i in range(-10, 0):
            if abs(i) > len(y_final_pred_test): continue
            pred_val = y_final_pred_test[i]
            actual_val = y_test2[i, target_idx]
            date = dates_test2[i]
            print(f"{date.strftime('%Y-%m-%d')} | Predicted = {pred_val:.2f}, Actual = {actual_val:.2f}")


    print("\n\n====================== Trading Signal for Tomorrow ======================")

    if 'Close' not in tomorrow_predictions:
        print("Could not generate a signal because the 'Close' price was not predicted.")
    else:
        last_known_close = df['Close'].iloc[-1]

        predicted_tomorrow_close = tomorrow_predictions['Close']

        TRADE_THRESHOLD = 0.025

        predicted_return = (predicted_tomorrow_close - last_known_close) / last_known_close

        signal = "Hold"
        reason = f"Predicted return of {predicted_return:.2%} is within the +/-{TRADE_THRESHOLD:.0%} threshold."

        if predicted_return > TRADE_THRESHOLD:
            signal = "Buy"
            reason = f"Model predicts the closing price will increase by {predicted_return:.2%}, which is above the {TRADE_THRESHOLD:.0%} buy threshold."
        elif predicted_return < -TRADE_THRESHOLD:
            signal = "Sell"
            reason = f"Model predicts the closing price will decrease by {abs(predicted_return):.2%}, which is beyond the -{TRADE_THRESHOLD:.0%} sell threshold."

        print(f"Last Known Close Price: {last_known_close:.2f}")
        print(f"Predicted Close Price for Tomorrow: {predicted_tomorrow_close:.2f}")
        print("--------------------------------------------------")
        print(f"ACTION:   {signal}")
        print(f"REASON:   {reason}")

    print("\n--- Full Prediction for Tomorrow ---")
    for target, pred_value in tomorrow_predictions.items():
        print(f"Predicted {target}: {pred_value:.2f}")


if __name__ == "__main__":
    main()
```
